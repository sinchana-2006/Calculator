```python
from google.colab import files
import time
import librosa
import numpy as np
import matplotlib.pyplot as plt
from scipy.fftpack import fft
from IPython.display import Audio

# Upload the MP3 file
uploaded = files.upload()

# Load the MP3 file (after uploading)
file_name = list(uploaded.keys())[0]  # Get the file name of the uploaded file
audio_data, sample_rate = librosa.load(file_name, sr=None)

# Play the first 20 seconds of the audio
duration = 20  # seconds
audio_20sec = audio_data[:duration * sample_rate]  # Extract the first 20 seconds of audio
audio_playback = Audio(audio_20sec, rate=sample_rate)

# Play the audio
display(audio_playback)

# Pause for 20 seconds while the audio is playing
time.sleep(duration)

# After the audio has finished playing, display the waveform and spectrum

# Display waveform for the first 20 seconds
plt.figure(figsize=(14, 5))
plt.plot(np.linspace(0, duration, len(audio_20sec)), audio_20sec)
plt.title('Waveform of the First 20 Seconds of the Audio File')
plt.xlabel('Time (seconds)')
plt.ylabel('Amplitude')
plt.show()

# Calculate and plot the spectrum (Fourier Transform) for the first 20 seconds
N = len(audio_20sec)
T = 1.0 / sample_rate
yf = fft(audio_20sec)
xf = np.linspace(0.0, 1.0/(2.0*T), N//2)

plt.figure(figsize=(14, 5))
plt.plot(xf, 2.0/N * np.abs(yf[:N//2]))
plt.title('Frequency Spectrum of the First 20 Seconds of the Audio File')
plt.xlabel('Frequency (Hz)')
plt.ylabel('Amplitude')
plt.grid()
plt.show()
```
